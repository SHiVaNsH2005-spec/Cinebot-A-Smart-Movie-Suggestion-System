```python
import ast  # to safely evaluate strings into Python objects

def convert(text):
    # Step 1: Check if it's already a list
    if type(text) == list:
        return text

    # Step 2: Try to convert string to list of dictionaries
    try:
        data = ast.literal_eval(text)  # Convert string to Python object
        result = []
        for item in data:
            result.append(item['name'])  # Get 'name' from each dictionary
        return result
    except:
        return []  # If there's an error, return an empty list
```

```python
movies['genres'] = movies['genres'].apply(convert)
movies.head()
```

```python
movies['keywords']=movies['keywords'].apply(convert)
movies.head()
```

```python
import ast

def main_cast(text):
    # Step 1: If it's already a list
    if isinstance(text, list):
        return [i['name'] if isinstance(i, dict) else i for i in text[:3]]

    # Step 2: Try to convert string to list of dicts
    try:
        data = ast.literal_eval(text)
        return [i['name'] for i in data[:3]]
    except:
        return []  # Return empty list on error
```

```python
movies['cast']=movies['cast'].apply(main_cast)
movies
```

```python
import ast

def fetch_director(text):
    L = []
    try:
        data = ast.literal_eval(text) if isinstance(text, str) else text
        for i in data:
            if isinstance(i, dict) and i.get('job') == 'Director':
                L.append(i['name'])
    except:
        pass
    return L
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
movies
```

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split()) #converting overview to list
```

```python
movies.head()
```

```python
def collapse(L):
    L1 = []
    for i in L:
        L1.append(i.replace(" ",""))
    return L1
```

```python
movies['cast'] = movies['cast'].apply(collapse)
movies['crew'] = movies['crew'].apply(collapse)
movies['genres'] = movies['genres'].apply(collapse)
movies['keywords'] = movies['keywords'].apply(collapse)
```

```python
movies.head()
```

```python
movies['tags']=movies['overview']+movies['genres']+movies['keywords']+movies['cast']+movies['crew']
movies['tags']
```

```python
import nltk
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    for i in text:
        i=i.lower()
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
movies['tags']=movies['tags'].apply(stem)
```

```python
movies['tags'][2]
```

```python
new = movies.drop(columns=['overview','genres','keywords','cast','crew'])
```

```python
new['tags'][0]
```

```python
new['tags'] = new['tags'].apply(lambda x: " ".join(x.split())).copy()
new.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000,stop_words='english')
vector=cv.fit_transform(new['tags']).toarray()
```

```python
vector
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity=cosine_similarity(vector)
similarity[0]
```

```python
mask = new['title'] == 'The Dark Knight Rises'
new[mask].index
```

```python
sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x:x[1])[1:6] #by this code we found nearest distance from first movie in descending order 
```

```python
def recommend(movie):
    mask = new['title'] == movie
    if not mask.any():
        return f"Movie '{movie}' not found in dataset."
    movie_index = new[mask].index[0]
    distance=similarity[movie_index]
    movies_list=sorted(list(enumerate(distance)),reverse=True,key=lambda x:x[1])[1:6]
    for i in movies_list:
        print (new.iloc[i[0]].title) # i[0] is the index 
```

```python
recommend('The Dark Knight Rises')
```
